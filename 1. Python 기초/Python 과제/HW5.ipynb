```python
class Set:
    def __init__(self, data=[]):
        for i in data:
            while data.count(i) != 1:
                data.remove(i)
                
        self.data = data
                
    
    def add(self, elem):
        if elem not in self.data:
            self.data.append(elem)
    
    
    def discard(self, elem):
        if elem in self.data:
            self.data.remove(elem)
    
    
    def clear(self):
        self.data.clear()
    
    
    def __len__(self):
        return len(self.data)

    
    def __str__(self):
        return '{' + ', '.join(map(str, self.data)) + '}'

    
    def __contains__(self, elem):
        if elem in self.data:
            return True
        else:
            return False
    

    def __le__(self, other):
        for i in self.data:
            if i not in other.data:
                return False
        return True
    
    
    def __ge__(self, other):
        for i in other.data:
            if i not in self.data:
                return False
        return True
    
    
    def __or__(self, other):
        data = self.data + other.data
        
        for i in data:
            while data.count(i) != 1:
                data.remove(i)
                
        return Set(data)
    
    
    def __and__(self, other):
        
        if len(self.data) >= len(other.data):
            data = [elem for elem in other.data if elem in self.data]
        else:
            data = [elem for elem in self.data if elem in other.data]
            
        return Set(data)
    
    
    def __sub__(self, other):
        data = [elem for elem in self.data if elem not in other.data]
        return Set(data)
    
    
    def __ior__(self, other):
        self.data = self.data + other.data
        
        for i in self.data:
            while self.data.count(i) != 1:
                self.data.remove(i)
                
        return self
    
    
    def __iand__(self, other):
         
        if len(self.data) >= len(other.data):
            self.data = [elem for elem in other.data if elem in self.data]
        else:
            self.data = [elem for elem in self.data if elem in other.data]        
        return self
    
    
    def __isub__(self, other):
        self.data = [elem for elem in self.data if elem not in other.data]        
        return self
```

```python
a = Set([1, 2, 3, 4])
b = Set([2, 3, 4])

print(a)
print(b)
print()

a.discard(4)
b.discard(2)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
```
