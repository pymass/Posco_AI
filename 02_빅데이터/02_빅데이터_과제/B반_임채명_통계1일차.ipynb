# 실습 문제 2

```python
for column in df.columns:
    if df[column].dtype == 'object':
        counts = df[column].value_counts().sort_index()
        cumulative_counts = np.cumsum(counts)
        percents = counts / sum(counts) * 100
        cumulative_percents = np.cumsum(percents)
        print('column name: {}'.format(column))
        print(counts)
        print(cumulative_counts)
        print(percents)
        print(cumulative_percents)
        print()
```

# 실습 문제 3

```python
mu = 175
sigma = 5
x = 180
prob = stats.norm.cdf(x, mu, sigma)
print('{}cm 이상인 비율: {:.1f}%'.format(x, (1-prob)*100))
```

# 실습 문제 4

```python
mu = 12
sigma = 3
x = 15
prob = stats.norm.cdf(x, mu, sigma)
print('{}km 이상 {}km 이하를 달릴 확률: {:.1f}%'.format((mu - sigma), (mu + sigma), (prob - 0.159)*100))
print('{}km 이상을 달릴 확률: {:.1f}%'.format(x, (1-prob)*100))
```

# 실습 문제 5

```python
t = 2.0
df = 100
prob = stats.t.cdf(t,df)
print('P(T<=F): {:.3f}'.format(prob))
```

# 실습 문제 6

```python
chisq = 3
df = 8
prob = stats.chi2.cdf(chisq, df)
print('P(X<={}) : {:.4f}'.format(chisq, prob))
```

# 실습 문제 7

```python
f = 1.5
dfnum = 10
dfden = 100

prob = stats.f.cdf(x=f, dfn = dfnum, dfd = dfden)
print("P(X<=F): {:.3f}".format(prob))
```

# 실습 문제 8

```python
x = 2500
alpha = 5.2
beta = 2200

prob = stats.weibull_min.cdf(x, alpha, scale=beta)
print("P(X<=x): {:.3f}".format(1-prob))
```

# 실습 문제 9

```python
n = 5
for i in range(n+1):
    prob = stats.binom.pmf(k = i, n = n, p = 0.3)
    print("P(X={}) = {:.3f}".format(i, prob))
```

# 실습 문제 10

```python
mu = 3
prob = stats.poisson.pmf(2, mu)
cdf_prob = stats.poisson.cdf(1, mu)
print("10시에서 11시 사이에 {}명의 손님이 도착할 확률:{:.2f}".format(2, prob))
print("10시에서 11시 사이에 {}명 이하의 손님이 도착할 확률:{:.2f}".format(1, cdf_prob))
print("10시에서 11시 사이에 {}명 이상의 손님이 도착할 확률:{:.2f}".format(3, 1 - (prob + cdf_prob)))
```
