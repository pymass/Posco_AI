# P.65 실습

## STEP1

```python
# cctv 데이터 불러오기
cctv_seoul = pd.read_excel('../Data/2. 데이터 Handling/서울시CCTV.xlsx')
cctv_seoul.head()
```

```python
# 컬럼명 변경
cctv_seoul.rename(columns={'기관명':'구별', '소계':'cctv수'}, inplace=True)
cctv_seoul.head()
```

```python
# 인구통계 데이터 불러오기
pop_seoul = pd.read_csv('../Data/2. 데이터 Handling/서울인구통계.txt', sep='\t', engine='python', thousands=',', encoding='utf-8')
pop_seoul.head()
```

```python
# 불필요 데이터 제거
pop1 = pop_seoul[['자치구', '인구', '인구.3', '인구.6', '65세이상고령자']]
pop_seoul = pop1.drop([0,1])
pop_seoul.head()
```

```python
# 컬럼명 변경
pop_seoul.columns = ['구별', '인구수', '한국인', '외국인', '고령자']
pop_seoul.head()
```

## STEP2

```python
cctv_seoul.sort_values(by='cctv수', ascending=True).head(10) # 오름차순
```

```python
cctv_seoul.sort_values(by='cctv수', ascending=False).head(10) # 내림차순
```

```python
# nan 값 개수 확인
cctv_seoul.isnull().sum()
```

```python
# 2011년 이전 열을 기준으로 nan값 추출
cctv_seoul.loc[cctv_seoul.isnull()['2011년 이전']]
```

```python
# 최근 3년 간 cctv 증가율 비교
cctv_seoul['최근증가율'] = (cctv_seoul['2018년'] + cctv_seoul['2017년'] + cctv_seoul['2016년']) / (cctv_seoul['cctv수'] - (cctv_seoul['2018년']+cctv_seoul['2017년']+cctv_seoul['2016년'])) * 100
cctv_seoul.sort_values(by='최근증가율', ascending=False).head()
```

## STEP3

```python
# 불필요 행 제거
pop_seoul.drop([2], inplace=True)
pop_seoul.head()
```

```python
# 컬럼별 데이터 타입 확인
pop_seoul.dtypes
```

```python
# 계산을 위해 int로 변경
pop_seoul = pop_seoul.astype({'인구수':'int', '한국인':'int', '외국인':'int', '고령자':'int'})
```

```python
# 외국인 비율과 고령자 비율 계산 및 추가
pop_seoul['외국인 비율'] = pop_seoul['외국인'] / pop_seoul['인구수'] * 100
pop_seoul['고령자 비율'] = pop_seoul['고령자'] / pop_seoul['인구수'] * 100
pop_seoul.head()
```

```python
pop_seoul.sort_values(by='외국인 비율', ascending=False).head() # 오름차순
```

```python
pop_seoul.sort_values(by='외국인 비율', ascending=True).head() # 내림차순
```

## STEP4

```python
# 공백 제거
cctv_seoul['구별'] = cctv_seoul['구별'].str.replace(' ', '')
```

```python
# 구를 기준으로 병합
result = pd.merge(cctv_seoul, pop_seoul, on='구별')
result
```

```python
# 불필요한 컬럼 제거
result.drop(columns=['2011년 이전', '2012년','2013년','2014년','2015년','2016년','2017년','2018년'], inplace=True)
```

```python
result.head()
```
